```python
%sql
SELECT DISTINCT language
FROM imdb.silver.silver_title_akas
WHERE language IS NOT NULL
ORDER BY language;
```

```python
%sql
-- Languages in silver but missing in gold
SELECT DISTINCT s.language
FROM imdb.silver.silver_title_akas s
LEFT JOIN imdb.gold.DIM_LANGUAGES g
  ON s.language = g.LANGUAGE_ABBR
WHERE s.language IS NOT NULL
  AND g.LANGUAGE_ABBR IS NULL;
```

```python
%sql
SELECT
    t.PRIMARY_TITLE,
    l.LANGUAGE_ABBR
FROM imdb.gold.BRIDGE_TITLE_LANGUAGE bl
JOIN imdb.gold.DIM_LANGUAGES l ON bl.LANGUAGE_SK = l.LANGUAGE_SK
JOIN imdb.gold.DIM_TITLES t     ON bl.TITLE_SK  = t.TITLE_SK
ORDER BY t.PRIMARY_TITLE
LIMIT 20;
```

```python
%sql
SELECT COUNT(*) AS null_langs
FROM imdb.gold.dim_languages
WHERE LANGUAGE_ABBR IS NULL;
```

```python
%sql
SELECT LANGUAGE_ABBR, COUNT(*)
FROM imdb.gold.dim_languages
GROUP BY LANGUAGE_ABBR
HAVING COUNT(*) > 1;
```
